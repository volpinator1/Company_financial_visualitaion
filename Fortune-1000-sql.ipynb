Before we begin our analysis we should verify that our data is complete and does not contain any null values or empty entries.

```python
%%sql

SELECT *
FROM companies
WHERE NOT (company IS NULL OR
rank IS NULL OR
rank_change IS NULL OR
revenue IS NULL OR
profit IS NULL OR
"num. of employees" IS NULL OR
sector IS NULL OR
city IS NULL OR
state IS NULL OR           
newcomer IS NULL OR           
ceo_founder IS NULL OR           
ceo_woman IS NULL OR           
profitable IS NULL OR
prev_rank IS NULL OR
CEO IS NULL OR           
Website IS NULL OR
Ticker IS NULL OR
"Market Cap" IS NULL                     
          IS NOT NULL)

##no NULL rows appeared
```

```python


%%sql

SELECT rank,company, city, profit
FROM companies
ORDER BY profit DESC
```

```python
%%sql

SELECT state, sum(revenue) AS TotalRevenue
FROM companies
GROUP BY state
ORDER BY TotalRevenue DESC
```

```python
%%sql

SELECT rank,company,"Market Cap", profit, sector
FROM companies
WHERE "Market Cap" IS NOT NULL
ORDER BY cast("Market Cap" as int) DESC;



#conclusiong: most high cap is tech, amzn can be cuonted as tech
```

```python


#expenses vs newcomer
```

```python
%%sql

SELECT AVG(revenue-profit) AS avg_expenses,AVG(profit) AS avg_profit, newcomer
FROM companies
GROUP BY newcomer
```

so, on average newcomers make a higher profit and also have bigger expenses. This shows us that they make more revenue than newcomer companies in our dataset.
